```python
# Step 2: Import Necessary Libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import re
```

```python
# Step 3: Load the Dataset from Google Drive
dataset_path = '/content/drive/MyDrive/Resume.csv'  # Update this with your file path
df = pd.read_csv(dataset_path)
```

```python
# Step 4: Input the Job Criteria
job_criteria = "Looking for a Python developer with experience in machine learning and data analysis. Skills in Python, TensorFlow, data analysis, and problem-solving are essential."
```

```python
# Step 5: Extract CV Texts
cv_texts = df['Resume_str'].tolist()  # Extract resume text
```

```python
# Step 6: Combine Job Criteria with CV Texts
documents = [job_criteria] + cv_texts  # Job criteria as the first document
```

```python
# Step 7: Compute TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
# Step 8: Calculate Cosine Similarity
similarities = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()
```

```python
# Step 9: Rank CVs by Similarity
cv_rankings = sorted(
    enumerate(similarities, start=1),
    key=lambda x: x[1],
    reverse=True
)
```

```python
# Step 10: Provide Specific Feedback for Each Ranked CV
def extract_keywords(text):
    """
    Extracts keywords from a text by looking for words with significant weight in TF-IDF.
    """
    # Extract words using regular expressions (ignoring stop words)
    words = re.findall(r'\b\w+\b', text.lower())
    return set(words)
```

```python
# Step 11: Provide detailed feedback for each ranked CV
print("Job Criteria:", job_criteria)
print("\nCV Rankings and Feedback:")

for rank, (cv_index, score) in enumerate(cv_rankings, start=1):
    cv_id = df.loc[cv_index - 1, 'ID']
    cv_text = df.loc[cv_index - 1, 'Resume_str']

    # Extract keywords from job criteria and CV
    job_keywords = extract_keywords(job_criteria)
    cv_keywords = extract_keywords(cv_text)

    # Find the matching keywords between job criteria and CV
    matched_keywords = job_keywords.intersection(cv_keywords)

    # Feedback based on the specific matches
    if len(matched_keywords) > 5:
        feedback = f"Strong match! The CV contains many relevant keywords such as {', '.join(matched_keywords)}."
    elif len(matched_keywords) > 2:
        feedback = f"Moderate match. The CV includes some key skills like {', '.join(matched_keywords)}."
    else:
        feedback = f"Low match. The CV lacks important skills mentioned in the job criteria."

    print(f"Rank {rank}: CV ID {cv_id} - Similarity Score: {score:.2f} - Feedback: {feedback}")
```
